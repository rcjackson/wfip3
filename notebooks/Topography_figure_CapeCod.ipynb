```python
import pygmt
import matplotlib.pyplot as plt
```

```python
minlat = 41.95
maxlat = 42.1
minlon = -70.3
maxlon = -69.9
highlands_lat_lon = (42.03265830998684, -70.05313755482639)
region = [minlon, maxlon, minlat, maxlat]
# Load sample grid (3 arc-minutes global relief) in target area
grid = pygmt.datasets.load_earth_relief(resolution="01s", region=region)
grid = pygmt.grdclip(grid, below=[0, 0])
fig = pygmt.Figure()
# make color pallets
pygmt.makecpt(
    cmap='oleron',
    series='0/70/1',
    continuous=True
)
#plot high res topography
fig.grdimage(
    grid=grid,
    region=region,
    projection='M4i',
    )
#fig.grdcontour(grid=grid, levels=50, region=region, annotation=10)
fig.coast(
    region=region,
    projection='M4i',
    frame=True,
    water='skyblue',
    shorelines='2p',
    )
fig.colorbar(
    frame='+lElevation [m ASL]'
    )
fig.plot(x=highlands_lat_lon[1], y=highlands_lat_lon[0], size=5, style="kstar4/0.3c", pen="1p,black", fill='black')
fig.text(x=highlands_lat_lon[1]+0.06, y=highlands_lat_lon[0], text='Highlands Center')
fig.basemap(map_scale="g-70.2/41.97+w5k")
fig.show()
```
