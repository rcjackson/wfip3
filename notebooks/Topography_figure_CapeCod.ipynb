```python
import pygmt
import matplotlib.pyplot as plt
```

```python
minlat = 40.8
maxlat = 42.3
minlon = -72
maxlon = -69.0
highlands_lat_lon = (42.03265830998684, -70.05313755482639)
nant_lat_lon = (41.2416792910947, -70.10714857877515)
bloc_lat_lon = (41.16674721495096, -71.58057127627573)
rhod_lat_lon = (41.4456, -71.4357)
mvco_lat_lon = (41.3499333, -70.52666667)
mvco2_lat_lon = (41.3619, -70.5223)


region = [minlon, maxlon, minlat, maxlat]
# Load sample grid (3 arc-minutes global relief) in target area
grid = pygmt.datasets.load_earth_relief(resolution="01s", region=region)
#grid = pygmt.grdclip(grid, below=[0, 0])
fig = pygmt.Figure()
# make color pallets
pygmt.makecpt(
    cmap='oleron',
    series='0/200/1',
    continuous=True
)
#plot high res topography
fig.grdimage(
    grid=grid,
    region=region,
    projection='M4i',
    )
#fig.grdcontour(grid=grid, levels=50, region=region, annotation=10)
fig.coast(
    region=region,
    projection='M4i',
    frame=True,
    water='skyblue',
    shorelines='2p',
    )
fig.colorbar(
    frame='+lElevation [m ASL]'
    )
fig.plot(x=highlands_lat_lon[1], y=highlands_lat_lon[0], 
         size=65, style="kstar4/0.3c", pen="1p,black", fill='yellow')
fig.text(x=highlands_lat_lon[1]+0.2, y=highlands_lat_lon[0], text="CACO")
fig.plot(x=nant_lat_lon[1], y=nant_lat_lon[0], 
         size=65, style="kstar4/0.3c", pen="1p,black", fill='yellow')
fig.text(x=nant_lat_lon[1]+0.2, y=nant_lat_lon[0], text="NANT")
fig.plot(x=bloc_lat_lon[1], y=bloc_lat_lon[0], 
         size=65, style="kstar4/0.3c", pen="1p,black", fill='yellow')
fig.text(x=bloc_lat_lon[1]+0.2, y=bloc_lat_lon[0], text="BLOC")
fig.plot(x=rhod_lat_lon[1], y=rhod_lat_lon[0], 
         size=65, style="kstar4/0.3c", pen="1p,black", fill='yellow')
fig.text(x=rhod_lat_lon[1]+0.2, y=rhod_lat_lon[0], text="RHOD")
fig.plot(x=mvco_lat_lon[1], y=mvco_lat_lon[0], 
         size=65, style="kstar4/0.3c", pen="1p,black", fill='yellow')
fig.text(x=mvco_lat_lon[1]+0.2, y=mvco_lat_lon[0], text="MVCO")
#fig.plot(x=mvco2_lat_lon[1], y=mvco2_lat_lon[0], 
#         size=55, style="kstar4/0.3c", pen="1p,red", fill='red')
#fig.text(x=highlands_lat_lon[1]+0.06, y=highlands_lat_lon[0], text='Highlands Center')
fig.basemap(map_scale="g-69.5/41.0+w50k")
fig.show()
fig.savefig('wfip3_site_fig.png')
```

```python
minlat = 45 + 30/60
maxlat = 46 
minlon = -122.35
maxlon = -121.4
wref_lat_lon = (45.81264444, -121.94989444)

region = [minlon, maxlon, minlat, maxlat]
# Load sample grid (3 arc-minutes global relief) in target area
grid = pygmt.datasets.load_earth_relief(resolution="01s", region=region)
#grid = pygmt.grdclip(grid, below=[0, 0])
fig = pygmt.Figure()
# make color pallets
pygmt.makecpt(
    cmap='oleron',
    series='0/1500/10',
    continuous=True
)
#plot high res topography
fig.grdimage(
    grid=grid,
    region=region,
    projection='M4i',
    )
#fig.grdcontour(grid=grid, levels=50, region=region, annotation=10)
fig.coast(
    region=region,
    resolution='f',
    projection='M4i',
    frame=True,
    water='skyblue',
    shorelines='2p',
    rivers="1/5p,blue"
    )
fig.colorbar(
    frame='+lElevation [m ASL]'
    )
fig.plot(x=wref_lat_lon[1], y=wref_lat_lon[0], 
         size=5, style="kstar4/0.3c", pen="1p,red", fill='yellow')
#fig.plot(x=mvco2_lat_lon[1], y=mvco2_lat_lon[0], 
#         size=55, style="kstar4/0.3c", pen="1p,red", fill='red')
fig.text(x=wref_lat_lon[1]+0.06, y=wref_lat_lon[0],  text='WREF')
fig.basemap(map_scale="g-121.6/45.55+w20k")
fig.show()
fig.savefig('leaff_site_fig.png')
```
